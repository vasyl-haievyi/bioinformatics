```python
from Bio import Entrez
Entrez.email = "vasyl.haievyi@student.uj.edu.pl"

handle = Entrez.esearch(db='gene', term="BRCA1", retmax=30)
result = Entrez.read(handle)

id_list = result['IdList']

handle = Entrez.esummary(db='gene', id=",".join(id_list))
result = Entrez.read(handle)
```

```python
BRCA1_id = None

for record, id in zip(result['DocumentSummarySet']['DocumentSummary'], id_list):
    keys = ["Name", "Chromosome", "MapLocation", "Description"]
    str = ", ".join([f"{key} : {record[key]}" for key in keys])
    print(str)

    if record['Name'] == 'BRCA1':
        BRCA1_id = id

assert(BRCA1_id is not None)
```

```python
handle = Entrez.elink(dbfrom='gene', db='omim', id=BRCA1_id)
result = Entrez.read(handle)

omim_id_list = list(map(lambda entry: entry['Id'], result[0]['LinkSetDb'][0]['Link']))

handle = Entrez.esummary(db='omim', id=",".join(omim_id_list))
result = Entrez.read(handle)
```

```python
for record in result:
    print(record['Title'])
```

Bazy danych posiadają dużą ilość informacji takich jak nazwe związku, sekwencję aminokwasów, strukturę przestrzenną etc.

Posiadając te informację możemy analizować podobięstwo różych bialek uwzgędniając charakterystyki jakościowe.
